```python
import pandas as pd
```

```python
df=pd.read_csv("superstore.csv",encoding="unicode_escape")
```

```python
df
```

```python
h=df.groupby("Sub-Category")["Sales"].sum()
h
```

```python
import matplotlib.pyplot as plt
x=[]
for i in range(len(h)):
    x.append(h.index[i])
y=h.to_list()
xpos=list(range(10,171,10))
plt.figure(figsize=(20,20))
plt.bar(xpos, y,width=5)
plt.xticks(xpos, x)
plt.show()
```
